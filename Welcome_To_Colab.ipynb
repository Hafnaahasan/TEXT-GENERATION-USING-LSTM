```
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

data = """
Artificial Intelligence is transforming the world rapidly.
Machine learning and deep learning techniques allow computers
to learn patterns from data and make intelligent decisions.
Natural Language Processing helps machines understand and
generate human language. LSTM networks are powerful models
used for sequence prediction tasks such as text generation,
speech recognition, and time series forecasting.

Technology continues to evolve and shape the future of society.
Researchers and engineers work together to create smarter
systems that assist humans in solving complex problems.
"""

print(data[:500])

chars = sorted(list(set(data)))
char_to_index = {c: i for i, c in enumerate(chars)}
index_to_char = {i: c for i, c in enumerate(chars)}

vocab_size = len(chars)
print("Total Characters:", vocab_size)

seq_length = 40
X = []
y = []

for i in range(len(data) - seq_length):
    seq_in = data[i:i + seq_length]
    seq_out = data[i + seq_length]
    X.append([char_to_index[char] for char in seq_in])
    y.append(char_to_index[seq_out])

n_patterns = len(X)
print("Total Patterns:", n_patterns)

X = np.reshape(X, (n_patterns, seq_length, 1))
X = X / float(vocab_size)

y = tf.keras.utils.to_categorical(y)

model = Sequential()
model.add(LSTM(256, input_shape=(X.shape[1], X.shape[2])))
model.add(Dense(vocab_size, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam')

model.summary()

model.fit(X, y, epochs=8, batch_size=64)

def sample(preds, temperature=0.7):
    preds = np.asarray(preds).astype('float64')
    preds = np.log(preds + 1e-8) / temperature
    exp_preds = np.exp(preds)
    preds = exp_preds / np.sum(exp_preds)
    return np.random.choice(len(preds), p=preds)

start = np.random.randint(0, len(X) - 1)
pattern = X[start]

generated_text = ""

for i in range(200):
    prediction = model.predict(pattern.reshape(1, seq_length, 1), verbose=0)[0]
    index = sample(prediction, temperature=0.7)
    result = index_to_char[index]
    generated_text += result
    pattern = np.append(pattern, index / vocab_size)
    pattern = pattern[1:]

print("Generated Text:")
print(generated_text)
```
